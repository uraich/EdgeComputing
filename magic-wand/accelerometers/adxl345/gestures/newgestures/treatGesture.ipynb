# Treat gesture data
* Read the gesture data
* Remove files that are too short
* Remove files that do not have inactivity at the end
* Shorten all valid files to a length of 128 values


```python
import os
files = os.listdir()
```

```python
for file in files:
    if file.find('.dat') != -1:
        infile = open(file,"r") 
        buffer = infile.readlines() 
        file_length = len(buffer) # This would give length of files.
        infile.close()
        if file_length < 128 :
            print("Short file: {}, length is {:d}".format(file,file_length))
        else :
            outfile= open("treated/"+file,"w")
            for line in range(128):
                outfile.write(buffer[line])
            outfile.close()
```
