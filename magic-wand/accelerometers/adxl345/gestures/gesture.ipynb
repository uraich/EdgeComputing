# First gesture studies


```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Read the zero gesture

```python
zero_df = pd.read_csv("zero.dat")
zero_df = zero_df[:20]
zero_df.plot()
```

```python
zero_df.plot.scatter("ax","ay")
```

```python
one_df = pd.read_csv("one.dat")
one_df = one_df[:40]
one_df.plot()
```

```python
one_df.plot.scatter("ax","ay")
```

```python
def plot_positions(df):
    vx = []
    vy = []
    sx = []
    sy = []
    sum= 0
    # integreate once to get the speed
    for i in range(len(df['ax'])) :
        sum += df['ax'][i]
        vx.append(sum)
    sum = 0
    # integrate a second time to get the position
    for i in range(len(vx)) :
        sum += vx[i]
        sx.append(sum)
    # same for y
    sum = 0
    for i in range(len(df['ay'])) :
        sum += df['ay'][i]
        vy.append(sum)
    sum = 0
    for i in range(len(vy)) :
        sum += vy[i]
        sy.append(sum)
    plt.scatter(sx,sy)
    plt.show()
```

```python
plot_positions(zero_df)
```

```python
plot_positions(one_df)
```

# Top Down movement


```python
top_down_df = pd.read_csv("top_down.dat")
top_down_df = top_down_df[:30]
top_down_df.plot()
```

```python
plot_positions(top_down_df)
```

```python
left_right_df = pd.read_csv("left_right.dat")
left_right_df = left_right_df[:40]
left_right_df.plot()
```

```python
plot_positions(left_right_df)
```
